```python
import numpy as np
import matplotlib.pyplot as plt

ph = np.load('Photon.npy')
ele = np.load('Electron.npy')
pi0 = np.load('Pizero.npy')
pipm = np.load('Piplusminus.npy')
```

```python
data = np.vstack((ph[:100000,:],ele[:100000,:],pi0[:100000,:],pipm[:100000,:]))
datatest = np.vstack((ph[100000:125000,:],ele[100000:125000,:],pi0[100000:125000,:],pipm[100000:125000,:]))
datavad = np.vstack((ph[125000:150000,:],ele[125000:150000,:],pi0[125000:150000,:],pipm[125000:150000,:]))
datavad.shape
```

```python
truez = data[:,2]
psb = data[:,20:83]
emb1 = data[:,83:134]
emb2 = data[:,134:197]
emb3 = data[:,197:260]                   
hab1 = data[:,260:]

# sum of all cells
cells_sum = np.sum(data[:,20:],axis=1).reshape(-1,1)

# sum of layers in phi
psb_sum = np.sum(psb.reshape(-1,7,9),axis=1)
emb1_sum = np.sum(emb1.reshape(-1,3,17),axis=1)
emb2_sum = np.sum(emb2.reshape(-1,7,9),axis=1)
emb3_sum = np.sum(emb3.reshape(-1,7,9),axis=1)
hab1_sum = np.sum(hab1.reshape(-1,7,9),axis=1)

# normalised sums
psb_nsum = psb_sum / cells_sum
emb1_nsum = emb1_sum / cells_sum
emb2_nsum = emb2_sum / cells_sum
emb3_nsum = emb3_sum / cells_sum
hab1_nsum = hab1_sum / cells_sum

# center eta of layers
psb_eta = data[:,6]
emb1_eta = data[:,8]
emb2_eta = data[:,10]
emb3_eta = data[:,12]
hab1_eta = data[:,14]
differ1 = psb_eta - emb1_eta
differ2 = emb1_eta - emb2_eta
differ3 = emb2_eta - emb3_eta
differ4 = emb3_eta - hab1_eta
```

```python
train = np.column_stack((psb_nsum,emb1_nsum,emb2_nsum,emb3_nsum,hab1_nsum,
                              psb_eta.reshape(-1,1),emb1_eta.reshape(-1,1),
                          emb2_eta.reshape(-1,1),emb3_eta.reshape(-1,1),hab1_eta.reshape(-1,1),
                              differ1.reshape(-1,1),differ2.reshape(-1,1),differ3.reshape(-1,1),differ4.reshape(-1,1)))

train_target = truez
```

```python
truezv = datavad[:,2]
psbv = datavad[:,20:83]
emb1v = datavad[:,83:134]
emb2v = datavad[:,134:197]
emb3v = datavad[:,197:260]                   
hab1v = datavad[:,260:]

# sum of all cells
cells_sumv = np.sum(datavad[:,20:],axis=1).reshape(-1,1)

# sum of layers in phi
psb_sumv = np.sum(psbv.reshape(-1,7,9),axis=1)
emb1_sumv = np.sum(emb1v.reshape(-1,3,17),axis=1)
emb2_sumv = np.sum(emb2v.reshape(-1,7,9),axis=1)
emb3_sumv = np.sum(emb3v.reshape(-1,7,9),axis=1)
hab1_sumv = np.sum(hab1v.reshape(-1,7,9),axis=1)

# normalised sums
psb_nsumv = psb_sumv / cells_sumv
emb1_nsumv = emb1_sumv / cells_sumv
emb2_nsumv = emb2_sumv / cells_sumv
emb3_nsumv = emb3_sumv / cells_sumv
hab1_nsumv = hab1_sumv / cells_sumv

# center eta of layers
psb_etav = datavad[:,6]
emb1_etav = datavad[:,8]
emb2_etav = datavad[:,10]
emb3_etav = datavad[:,12]
hab1_etav = datavad[:,14]
differ1v = psb_etav - emb1_etav
differ2v = emb1_etav - emb2_etav
differ3v = emb2_etav - emb3_etav
differ4v = emb3_etav - hab1_etav
```

```python
vad = np.column_stack((psb_nsumv,emb1_nsumv,emb2_nsumv,emb3_nsumv,hab1_nsumv,
                              psb_etav.reshape(-1,1),emb1_etav.reshape(-1,1),
                          emb2_etav.reshape(-1,1),emb3_etav.reshape(-1,1),hab1_etav.reshape(-1,1),
                              differ1v.reshape(-1,1),differ2v.reshape(-1,1),differ3v.reshape(-1,1),differ4v.reshape(-1,1)))

vad_target = truezv
```

```python
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb

learning_rates = np.arange(0.1, 0.9, 0.02)
validation_errors = []
train_errors=[]

for lr in learning_rates:
    model = xgb.XGBRegressor(
        learning_rate=lr,
        n_estimators=600,
        max_depth=6,
        objective='reg:squarederror'
    )
    
    model.fit(
        train,train_target,
        eval_set=[(vad, vad_target)],
        early_stopping_rounds=50,
        verbose=False
    )
    
    # Calculate training RMSE
    train_preds = model.predict(train)
    train_rmse = np.sqrt(np.mean((train_target - train_preds) ** 2))
    
    # using RMSE as the performance metric
    validation_rmse = model.evals_result()['validation_0']['rmse'][-1]
    
    validation_errors.append(validation_rmse)
    train_errors.append(train_rmse)
```

```python
# Plot the validation RMSE for each learning rate
plt.plot(learning_rates, validation_errors, label='Validation Error')
plt.plot(learning_rates, train_errors, label='Training Error')
plt.title('Training and Validation RMSE by Learning Rate of Large Model with 62 inputs')
plt.xlabel('Learning Rate')
plt.ylabel('RMSE of the model (mm)')
plt.legend()
plt.grid(True)
plt.show()
```
